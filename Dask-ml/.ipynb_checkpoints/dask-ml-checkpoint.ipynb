# Data Prepocessing

The first process step in building a machine learning model is data cleaning. The data we have here is not very complex which makes data cleaning easier. In production quality ML model this is the most time consuming step. 

Data cleaning maily involves:
1. Remove any unnecessary observations from your dataset
2. Remove redundant informations
3. Remove duplicate informations
4. Remove structural errors in data collection
5. Remove unwanted outliers - outliers can result in overfitting
6. Handle missing data:
    * Remove observations with values missing
    * Infer the missing values

In this case we are taking the easiest method to address missing values. We are removing any dataframe row that has missing values. This is not always advicable as we are lossing a lot of information and in the end we end up not getting the entire picture.

Inferring data is also not always a good ideas as we maybe adding some bias to the inference.

Data usually have numerical data and catogorical data. 
1. Categorical data groups information (usuallytext) with similar characteristics 
2. Numerical data expresses information in the form of numbers

Ideally all catogorical data should be converted to numerical data. In this case we remove some all catogorical data other than 'RainToday' and 'RainTomorrow'.

## Corelation Matrix

Correlation is often used in machine learning to identify multicollinearity, which is when two or more predictor variables are highly correlated with each other. Multicollinearity can adversily affect the accuracy of predictive models.

* The coefficients become very sensitive to small changes in the model.
* Multicollinearity reduces the precision of the estimated coefficients, which weakens the statistical power of your regression model. 

Multicollinearity can be addressed removing one of the correlated variables 

```python
corr_matrix = ddf_features_norm.corr(method='pearson', min_periods=None, numeric_only='__no_default__', split_every=False)
```

```python
corr_matrix.compute()
```

```python

f, ax = plt.subplots(figsize=(12, 10))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr_matrix, annot=True, mask = mask, cmap=cmap)
```

Ideally we should remove one of the higly correlated feature or combine those together. For the time being we doing neither.

# Cross validation
Cross-validation is a method for evaluating ML models by training several ML models on subsets of the data and evaluating anonther subset of the data. The adavantages of cross validation are : 

1. Identify Overfitting
2. Comparisson between different models 
3. Hyperparameter tuning
4. Efficienncy : Allows the use of data for both training and validation

# Linear Regression

Linear regression tries to find the optimal W<sub>1</sub>, W<sub>2</sub>, W<sub>3</sub>, W<sub>4</sub>, so that we can predict the value of Y for the user supplied X<sub>1</sub>, X<sub>2</sub>, X<sub>3</sub>.

$$
  Y(X_1, X_2, X_3) = W_1 * X_1 + W_2 * X_2 + W_3 * X_3 + W_4
$$



```python
lr = LinearRegression(solver_kwargs={"normalize":False})
```

```python
lr.fit(X_train_array, y_train_array)
```

```python
y_test_array = y_test.to_dask_array(lengths=True)
X_test_array = X_test.to_dask_array(lengths=True)
```

```python
print(lr.score(X_test_array, y_test_array))
```

# K-Means Clustering

```python
from sklearn.datasets import fetch_california_housing

data = fetch_california_housing()
X, y = data.data, data.target
feature_names = data.feature_names


```

```python
mat = np.column_stack((X, y))
df = pd.DataFrame(mat, columns=np.append(feature_names, 'MedianValue'))
df.shape
```

```python
ddf = dd.from_pandas(df, npartitions=10).dropna()
ddf.head()
```

```python
sns.scatterplot(data = df, x = 'Longitude', y = 'Latitude', hue = 'MedianValue')
```

```python
ddf_features = ddf.drop(columns=['MedianValue'])
ddf_target = ddf[['MedianValue']]

X_train, X_test, y_train, y_test = train_test_split(ddf_features, ddf_target, test_size=0.2)
```

```python
pipe = Pipeline([('reduce_dim', PCA()), 
                 ('cluster', KMeans(n_clusters = 3, random_state = 0, n_init='auto'))
                ])
```

```python
kmeans = pipe.fit(X_train.to_dask_array(lengths=True))
```

```python
pipe['cluster'].labels_.compute()
```

```python
train_score = pipeline.score(X_train, y_train)
print('R2 score on the training set:', np.round(train_score, 5))

test_score = pipeline.score(X_test, y_test)
print('R2 score on the test set:', np.round(test_score, 5))
```

```python
sns.scatterplot(data = X_train, x = 'Longitude', y = 'Latitude', hue = pipe['cluster'].labels_)
```
